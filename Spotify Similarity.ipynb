```python
for col_name in df.columns:
    if(col_name == "track_id"):
        continue
    if(df[col_name].dtype == 'object'):
        df[col_name]= df[col_name].astype('category')
        df[col_name] = df[col_name].cat.codes
```

```python
import numpy as np
df1 = df[df.index.str.startswith(our_id)]
diff_df = df - df1.values
# or diff_df = df2 - df1.iloc[0, :]
norm_df = diff_df.apply(np.linalg.norm, axis=1)

```

```python
clean_df = pd.read_csv(r'spotify_dataset.csv')
print(clean_df)
clean_df = clean_df.drop_duplicates(subset=['track_id'])
print(clean_df)
```

```python

print(norm_df.nsmallest())
sorted_df = norm_df.nsmallest()
sorted_df.iteritems()
ids = []
for i, j in sorted_df.items():
    ids.append(i)
    
clean_df.loc[clean_df['track_id'].isin(ids)]
```
